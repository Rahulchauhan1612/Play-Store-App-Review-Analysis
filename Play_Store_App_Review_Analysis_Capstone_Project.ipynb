Now let's change the columns in not_null_df1 which are not in right format.

```python
#changing string type Reviews column to int64.

not_null_df2['Reviews'] = not_null_df2['Reviews'].astype('int64')
```

```python
#changing string type last updated column to datetime64.

not_null_df2['Last Updated'] = not_null_df2['Last Updated'].astype('datetime64[ns]')
```

```python
#Replacing '+' and ',' from installs column and converting it to int using eval.

def convertion_to_int(row):
    row=row.replace(',','')
    row=row.replace('+','')
    row=eval(row)
    
    return row
```

```python
not_null_df2['Installs'] = not_null_df2['Installs'].apply(convertion_to_int)
```

```python
not_null_df2.head()
```

```python
#Replacing the elements of size and converting it to 'KB' by multiplying the 'MB' value with 1024 by using eval function which returns int after parsing string.

def convert_size_to_string(row):
      row=row.lower()

      if row=='varies with device':
        row=0

      else:

        if 'm' in row:
          row=row.replace('m','')
          row=eval(row)*1024

        elif 'k' in row:
          row=row.replace('k','')
          row=eval(row)
          
      return row
```

```python
#Creating a new column named Size_in_KB to have all Size column value in KB.

not_null_df2['Size_in_KB'] = not_null_df2['Size'].apply(convert_size_to_string)
```
